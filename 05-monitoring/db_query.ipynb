```python
import sqlite3

conn = sqlite3.connect("traces.db")
cur = conn.cursor()


result = cur.execute("""
    SELECT * FROM spans
    """)

row_list = []
for row in cur.fetchall():
    row_list.append(row)
    print(row)

columns = [desc[0] for desc in cur.description]
```

```python
import pandas as pd

table_df = pd.DataFrame(row_list, columns=columns)
table_df
```

```python
filter_llm = table_df["name"] == "llm"
filter_search = table_df["name"] == "search"

llm_time = table_df[filter_llm]["end_time"] - table_df[filter_llm]["start_time"] 
search_time = table_df[filter_search]["end_time"] - table_df[filter_search]["start_time"]

print("The time difference:")
print(f"LLM: {llm_time.mean()}")
print(f"Search: {search_time.mean()}")
```
